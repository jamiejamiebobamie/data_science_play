```python
from collections import Counter
```

```python
crimes_histogram = Counter(seattle_CRIME_df['Crime Subcategory'])
```

```python
crimes_histogram
```

```python
import heapq
```

```python
keys_lookup = list(crimes_histogram.keys())
values_lookup = list(crimes_histogram.values())
print(len(keys_lookup))
print(len(values_lookup))

LOOKUP = {}
for i, value in enumerate(values_lookup):
    LOOKUP[value] = keys_lookup[i]
```

```python
top_five_crimes_amuounts = heapq.nlargest(5,values_lookup)
```

```python
top_five_crimes = []
for amount in top_five_crimes_amuounts:
    top_five_crimes.append((LOOKUP[amount], amount))
print(top_five_crimes)
```

### top five crimes by incident report

```python
x_val = [x[0] for x in top_five_crimes]
y_val = [x[1] for x in top_five_crimes]

print(x_val)
plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
plt.show()
```

## The West precinct followed by the North precinct handles the most violent crimes.

```python
violent_crimes_by_neighborhood_histogram = Counter(violent_crimes_df['Neighborhood'])
```

```python
violent_crimes_by_neighborhood_histogram

name_lookup = {}

for key, value in violent_crimes_by_neighborhood_histogram.items():
    name_lookup[value] = key
```

```python
top5_most_violent_areasOfSeattle = heapq.nlargest(5,violent_crimes_by_neighborhood_histogram.values())
```

```python
graph_data = []
for value in top5_most_violent_areasOfSeattle:
    print(name_lookup[value], value)
    graph_data.append((name_lookup[value], value))
```

```python
x_val = [x[0] for x in graph_data]
y_val = [x[1] for x in graph_data]

print(x_val)
plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
plt.show()
```
